```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'Data/実験2.txt'
# path = 'Data/実験3.txt'
# path = 'Data/実験4.txt'
```

```python
columns = ["nSeq", "I1", "I2", "O1", "O2", "A1", "A2", "A3", "A4", "A5", "A6"]
sampling_rate = 1000

df = None
reader = pd.read_csv(path, sep='\t', header=None, names=columns, index_col=False, chunksize=600*sampling_rate)
for r in reader:
    if df is None:
        #  頭から3行目は要らない
        df = r.drop([i for i in range(3)])
    else:
        df = df.append(r, ignore_index=True)
        
df = df.reset_index(drop=True)

print(df.shape)
df.head()
```

```python
def create_plot(df, title):
    if title:
        plt.title(title)
    plt.ylim([df.min(), df.max()])
    plt.vlines(60*9+11, df.min(), df.max(), 'red', linestyles='dashed', lw=3)
    plt.plot(df.index/sampling_rate, df)
```

```python
# timeの数値はsecond指定
start_time = int(60 * 9) # 実験2
# start_time = int(60 * 6.25) # 実験3
# start_time = int(60 * 6.2) # 実験4
time_interval = 20
end_time = start_time+time_interval

def create_plot_split(df, title=None):    
#     if title:
#         plt.title(title)
    seconds = np.arange(start_time, end_time, 1/sampling_rate)
    df = df[start_time*sampling_rate:start_time*sampling_rate+time_interval*sampling_rate]
    plt.ylim([df.min(), df.max()])
#     plt.ylim([470, 550])
    plt.plot(seconds, df)
    plt.vlines(60*9+11, df.min(), df.max(), 'red', linestyles='dashed', lw=3)
```

```python
df_emg = df['A1']
df_ecg = df['A2']
df_eda = df['A3']
```

```python
create_plot_split(df_emg, 'emg')
```

```python
create_plot_split(df_ecg, 'ecg')
```

```python
create_plot_split(df_eda, 'eda')
```

```python
create_plot(df_emg, 'emg')
```

```python
create_plot(df_ecg, 'ecg')
```

```python
create_plot(df_eda, 'eda')
```
